# Length of Stay vs Visit Amount Distribution

```python
def outlier_univariate_analysis(data, feature, figsize=(12,7), kde=False, bins=None):
    f2, (ax_box2, ax_hist2) = plt.subplots(nrows=2, sharex = True, gridspec_kw = {"height_ratios": (0.25, 0.75)}, figsize=figsize)
    sns.boxplot(data=data, x=feature, ax=ax_box2, showmeans=True, color='violet')
    sns.histplot(data=data, x = feature, kde=kde, ax=ax_hist2, bins=bins, palette='winter') if bins else sns.histplot(
    data=data, x=feature, kde=kde, ax=ax_hist2)
    ax_hist2.axvline(
    data[feature].mean(), color="green", linestyle="--")
    ax_hist2.axvline(
    data[feature].median(), color="black", linestyle="-")
```

```python
outlier_univariate_analysis(data, 'visit_amount')
```

```python
outlier_univariate_analysis(data, 'length_of_stay')
```

```python
fig = plt.figure(figsize=(18,6))
ax1 = fig.add_subplot(1,2,1)
data_temp =data[data.visit_amount < 20000]
sns.distplot(data_temp.loc[data.readmission_flag==1].visit_amount, bins = 20)
sns.distplot(data_temp.loc[data.readmission_flag==0].visit_amount, bins = 20)
plt.legend(["Readmission Visit Cost", "Admission Visit Cost"])
plt.title("Distribution of Visit Amount")
plt.xlim(-500, 20000)

ax2 = fig.add_subplot(1,2,2)
data_temp= data[data.length_of_stay < 15]
sns.distplot(data_temp.loc[data.readmission_flag==1].length_of_stay, bins = 10)
sns.distplot(data_temp.loc[data.readmission_flag==0].length_of_stay, bins = 10)
plt.legend(["Readmission LOS", " Admission Length of Stay"])
plt.title("Distribution of Length_of_stay")
plt.xlim(-1, 30)

plt.show()
```

Observation:
    1. The Distributions for the visit amount for admission vs readmission visits are different  with latters distribution being shifted towards left side. 
    2. Higher spread of cost is involed for readmitted patient 
    3. For Length of stay in case of admission vs readmission both are similar

# Based upon Chronic Disease, what is the inpatient readmission trend ?

```python
data_temp = data.groupby('disease_count').aggregate({'readmission_flag':['mean', 'count', 'sum']})[:20]
```

```python
READMISSION_POP_MEAN = round(data.readmission_flag.mean(), 3)
```

```python
fig  = plt.figure(figsize=(20,6))
ax1 = fig.add_subplot(1,1,1)
ax1.text(s = "# of Chornic Disease vs IP Reamdission", x =5, y = 2200, weight= 'bold', alpha=0.7, color='black', fontsize= 20)
ax1.bar(x = data_temp.index, height = data_temp[("readmission_flag", "count")], alpha=0.8, color='orange')
ax1.text(s = "# Ip Visit", x =-2.5, y = 550, alpha=0.7, color='black', fontsize= 20, rotation=90)
ax1.set_xlabel("# of chronic Diseases", fontsize=15, alpha=0.7, color='black')
ax1.text(s = "IP Readmission Proportion in %", x = 22, y = 350, alpha=0.7, color='black', fontsize= 20, rotation=90)
plt.grid()

ax2 = ax1.twinx()
ax2.plot(data_temp.index, 100*data_temp[("readmission_flag", "mean")], alpha=0.7, color = 'blue', marker='o', linestyle='dashed', 
        linewidth=2, markersize=7)
ax2.text(s=f'mean = {100*READMISSION_POP_MEAN}%', x = 0, y = READMISSION_POP_MEAN*100, fontsize=20, alpha=0.7, color='black')
plt.grid()

plt.legend(['IP Readmission Proportion'], loc = 'upper right', bbox_to_anchor=(0.93, 0.97))
plt.axhline(READMISSION_POP_MEAN*100, alpha=0.3, linewidth=2, linestyle="-.", color='black')

plt.show()
```

Observation:
    1. The above Analysis is done on 20 chronic diseasse obverserved in comorbid condition 
    2. clear trend is observed based upon increase in number of chronic disease 
    3. patient with more number of disease has higher side of readmission
